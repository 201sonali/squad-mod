3. Partial SQuAD, Synonym, and Numeric Data

```python
pip install datasets
```

```python
# LOADING ORIGINAL SQUAD DATA
from datasets import load_dataset
squad_dataset = load_dataset("squad", split='train')
```

```python
#CHECKING FORMAT
print("Number of examples in the 'train' split:", len(squad_dataset))

# Print details of the first 10 training examples
for i in range(10):
    print(f"Example {i+1}:")
    print("Question:", squad_dataset['question'][i])
    print("Context:", squad_dataset['context'][i])
    print("Answer:", squad_dataset['answers'][i])
    print("\n")
```

```python
# CONVERTING SQUAD DATASET TO JSON FOR COMBO

squad_dataset = squad_dataset.shuffle()

squad_exs = []

for i in range(10000):
    # checkpoing for iterations
    if (i % 5000) == 0:
      print(f"Iterations complete:", i)
    example = {'question': squad_dataset['question'][i], 'context': squad_dataset['context'][i], 'answers': squad_dataset['answers'][i]}
    squad_exs.append(example)
```

```python
print(len(squad_exs))
```

```python
# SAVING

# saving file as a json to add to training
import json

with open('/content/squad_exs.json', 'w') as json_file:
    json.dump(squad_exs, json_file, indent=2)
```

```python
# COMBINE JSONS FILES

import json
import random

json_file_paths = [
    "/content/fp-dataset-artifacts/adv_data/num_exs.json",
    "/content/fp-dataset-artifacts/adv_data/syn_exs_edited.json",
    "/content/train-v2.0.json",
    #"/content/fp-dataset-artifacts/adv_data/squad_exs.json",
]

combined_json_path = "/content/fp-dataset-artifacts/adv_data/combo_exs.json"

combo_data = []

for file_path in json_file_paths:
    with open(file_path, 'r') as json_file:
        file_data = json.load(json_file)
        combo_data.extend(file_data)

print(len(combo_data))

random.shuffle(combo_data)
print(len(combo_data))

with open(combined_json_path, 'w') as combined_json_file:
    json.dump(combo_data, combined_json_file, indent=2)
```
